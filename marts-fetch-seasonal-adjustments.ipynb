# Import seasonal adjustment factors

Source: https://www.census.gov/retail/mrts/www/nadjusts.txt

The government reports seasonal adjustment factors in a format that looks good on a green terminal screen but is fucking useless for processing. We have to jump through hoops to read it into a dataframe. Time for a revolution. We threw the redcoats out for less.

```python
import pandas as pd
import datetime
import dateutil
```

```python
months = []
for n in range(17):
    month = datetime.datetime(2018,12,31) + dateutil.relativedelta.relativedelta(months=n)
    month_abbrev = month.strftime('%Y-%m-%d')
    months.append(month_abbrev)

months
```

```python
# Specifications for the fixed width format
col_widths = (17, 51, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8, 8)
col_names = ['NAICS', 'Sector'] + months
```

```python
with open('nadjusts.txt', 'r') as fh:
    seas_adj = pd.read_fwf(
        fh, widths=col_widths, header=None, names=col_names, skiprows=4, skipfooter=12
    ).dropna(how='all').drop('NAICS', axis=1)
```

```python
seas_adj.to_csv('seas_adj.csv', index=False)
```
