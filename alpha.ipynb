```python
import snowflake.connector

def execute_snowflake_query(query, conn):
    ctx = snowflake.connector.connect(**conn)
    cur = ctx.cursor()
    cur.execute(query)
    df = cur.fetch_pandas_all()
    ctx.close()
    return df
```

```python
def pretty_format(df, index=None):
    d = df.copy()

    d.columns = map(str.lower, d.columns)

    if 'net_amt' in d.columns:
        d['net_amt'] = d['net_amt'].astype(float)

    date_columns = ['txn_date', 'tran_month']
    for dt in date_columns:
        if dt in d.columns:
            d[dt] = pd.to_datetime(d[dt])

    if index:
        d = d.set_index(index).sort_index()
    return d
```

# Improve forecast by adding features

```python
dataset = monthly.copy()
```

```python
dataset['avg_spend_per_card'] = dataset['net_amt'] / dataset['card_cnt']
dataset['avg_txn_amt'] = dataset['net_amt'] / dataset['txn_cnt']
dataset = dataset.merge(marts.shift(12).rename('marts_lagged_12m'), how='left', left_index=True, right_index=True)
```

```python
onehot = pd.get_dummies(dataset.index.month)
onehot.columns = dataset[:12].index.strftime('%b')
onehot.index = dataset.index
dataset = pd.concat([dataset, onehot], axis=1).sort_index()
```

```python
preds = {}
for month in dataset.index[-15:]:
    # train test split as a walkforward
    X_train = dataset.loc['2017':month+pd.offsets.MonthEnd(-1)]
    X_test = dataset.loc[month]
    y_train = marts.loc['2017':month+pd.offsets.MonthEnd(-1)]
    y_test = marts.loc[month]
    
    # scale data
    scaler = StandardScaler().fit(X_train)
    X_train_scaled = scaler.transform(X_train)
    X_test_scaled = scaler.transform(X_test.to_numpy().reshape(1,-1))
    
    # train the model
    lm = linear_model.LassoCV(cv=5, max_iter=1e4)
    lm.fit(X=X_train_scaled, y=y_train)
    
    
    prediction = {'y_pred': lm.predict(X=X_test_scaled)[0],
                 'y_true': y_test}
    preds[month.strftime('%Y-%m-%d')] = prediction
```

```python
preds
```

```python
y_pred_list = [preds[m]['y_pred'] for m in preds]
y_true_list = [preds[m]['y_true'] for m in preds]
```

```python
sqrt(mean_squared_error(y_true_list, y_pred_list))
```

## arm_prod_type

```python
prod_type_query = """
select
  tran_month,
  arm_prod_type,
  sum(txn_amt - coalesce(cash_back_amt, 0)) as net_amt,
  hll(arm_id_nbr) as card_cnt,
  hll(arm_tran_id) as txn_cnt

from
  alpha_full_csv_all as alpha
  
where
  arm_txn_cde in ('Spend', 'Non Financial', 'ATM Withdrawal')
  
group by
  tran_month,
  arm_prod_type

"""
```

```python
prod_type_raw = execute_snowflake_query(prod_type_query, alpha_connection)
```

```python
prod_type = pretty_format(prod_type_raw, index='tran_month')
```

```python
def transform_data(df_orig, category):
    df = df_orig.copy()
    df = df.loc['2012':'2020-04']
    df.index = df.index + pd.offsets.MonthEnd(1)
    
    df['avg_txn_size'] = df['net_amt'] / df['txn_cnt']
    df['avg_spend_per_card'] = df['net_amt'] / df['card_cnt']
    df['avg_txns_per_card'] = df['txn_cnt'] / df['card_cnt']
    
    df = df.reset_index().set_index([category, 'tran_month']).sort_index()
    return df
```

```python
prod_type = transform_data(prod_type, 'arm_prod_type')
```

```python
def collect_correlations(df):
    correlations = {}
    for cat in df.index.get_level_values(0).unique():
        cat_corrs = {}
        category = df.loc[cat]
        for col in category.columns:
            cat_corrs[col] = category[col].corr(marts)
        correlations[cat] = cat_corrs
    
    corrs = pd.DataFrame(correlations).T.sort_index()
    return corrs
```

```python
prod_type_corrs = collect_correlations(prod_type)
prod_type_corrs.sort_values(by='net_amt', ascending=False)
```

# Exponentially weighted correlations

This section can be skipped. TLDR: weighting the samples exponentially, with higher weight given to more recent data, has no discernible impact on predictive power.

```python
def w_mean(x, w):
    return np.average(x, weights=w)

def w_cov(x, y, w):
    return np.sum(w*(x-w_mean(x, w))*(y-w_mean(y, w))) / np.sum(w)

def w_corr(x, y, w):
    return w_cov(x, y, w) / np.sqrt(w_cov(x, x, w) * w_cov(y, y, w))
```

```python
def collect_w_correlations(df, w):
    correlations = {}
    for cat in df.index.get_level_values(0).unique():
        cat_corrs = {}
        category = df.loc[cat]
        for col in category.columns:
            cat_corrs[col] = w_corr(category[col], marts.loc['2012':'2020-04'], w)
        correlations[cat] = cat_corrs
    
    corrs = pd.DataFrame(correlations).T.sort_index()
    return corrs
```

# Conclusion

Suggested procedure for processing Alpha data:

* Filter transactions
    * `arm_txn_cde`: Spend, Non Financial, ATM Withdrawal
    * `arm_prod_type`: Corporate Expense, Incentive, Payroll, Government, Debit
    * `net_amt`: transaction below $900
    
* Reweight transactions by ZCTA populations
